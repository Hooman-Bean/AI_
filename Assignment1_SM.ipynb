```python
import pandas as pd
from csv import reader
import re
```

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
# load csv file
d = pd.read_csv('chipotle.csv')
d
```

```python
item_price=[]
for row in reader(open('chipotle.csv', 'r')):
    item_price.append(row[5])
    item_price = re.findall(r'\d+\.\d+', str(item_price))
    item_price = [float(i) for i in item_price] 
c=0
for i in range(len(item_price)):
    if item_price[i] > 10:
        c = c + 1
print(c, 'products cost more than $10.00')
```

```python
d[['item_name','item_price']].sort_values('item_name')
```

```python
print('The quantity of most expensive item oredered is', d['quantity'][d['Unnamed: 0'] == item_price.index(max(item_price))].values[0])
```

```python
print('Veggie Salad Bowl was ordered',d[d['item_name'] == 'Veggie Salad Bowl'].count().values[0],'times')
```

```python
print('More than one Canned Soda was ordered',d[d['item_name'] == 'Canned Soda'][d['quantity'] > 1].count().values[0],'times')
```

```python
data1=pd.DataFrame(raw_data_1)
data2=pd.DataFrame(raw_data_2)
data3=pd.DataFrame(raw_data_3)
```

```python
all_data=pd.merge(data1,data2,how='outer',on='subject_id')
all_data
```

```python
all_data_col=pd.merge(data1,data2,how='outer')
all_data_col
```

```python
all_data_data_3=pd.merge(all_data,data3,how='outer',on='subject_id')
all_data_data_3
```

```python
same_subjectID_data1_data2=pd.merge(data1,data2,how='inner',on='subject_id')
same_subjectID_data1_data2
```

```python
all_values_data1_data2=pd.merge(data1,data2,how='outer',on=['subject_id','first_name','last_name'])
all_values_data1_data2
```
